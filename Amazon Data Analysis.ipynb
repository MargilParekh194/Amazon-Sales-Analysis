```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
# importing Data
file_path = 'amazon-sales-dataset.txt'
df = pd.read_csv(file_path)
```

```python
# previewing data 
print(df.shape)
df.head()
```

Begin data clening process

```python
# check datatypes and null values
df.info()
df.isnull().sum()
```

```python
# Clean the price columns - actual and discounted price columns
def clean_price(col):
    return df[col].str.replace('₹', '').str.replace(',', '').astype(float)
df['discounted_price'] = clean_price('discounted_price')
df['actual_price'] = clean_price('actual_price')
```

```python
df.head()
```

```python
# clean the rating_count column and convert it to float
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Split the category column
categories = df['category'].str.split('|', expand=True)
df[['category_main','category_sub1','category_sub2']] = categories.iloc[:, :3]
df.drop(columns=['category'], inplace=True)
df.head()
```

```python
# Begin exploratory data analysis
# Basic statistics

df.describe()
```

```python
# Display the top 10 most expensive products 
df.sort_values('actual_price', ascending = False)[['product_name','category_main','actual_price']].head(10)
```

```python
# Calculate and display the best selling category
category_sum = df.groupby('category_main')['actual_price'].sum().sort_values(ascending = False)
category_sum.plot(kind='bar', figsize=(10,6), color='green')
plt.title('Best selling categories')
plt.xlabel('category')
plt.ylabel('Revenue')
plt.show()
```

```python
# Best Rated category
category_rating = df.groupby('category_main')['rating'].mean().sort_values(ascending = False)
category_rating.plot(kind='bar', figsize=(10,6), color = 'orange')
plt.title('Best Rated category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```
